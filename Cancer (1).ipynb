**Model**

```python
# Libraries for model building
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Library for stat model
import statsmodels.api as sm

# Library for checking multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

import warnings
warnings.filterwarnings('ignore')
```

```python
final_df = df.rename(columns={'avganncount':'avg_count', 'avgdeathsperyear':'avg_peryear', 'incidencerate':'incident_rate', 'target_deathrate':'target_death'})[['avg_count','avg_peryear','incident_rate','target_death']]
```

```python
X = final_df.iloc[:, :-1]
Y = final_df.iloc[:, -1]

# Defining the modell
X = sm.add_constant(X)
model = sm.OLS(Y, X)

# Fitting the model
result = model.fit()

# Printing the model summary
print(result.summary())
```

```python
plt.rcParams["figure.figsize"] = (20,5)

plt.subplot(1, 3, 1)
plt.title("Avarge Age wise")
sns.regplot(X['avg_count'], Y, scatter_kws={"color": "cyan"}, line_kws={"color": "red"})

plt.subplot(1, 3, 2)
plt.title("Per Year")
sns.regplot(X['avg_peryear'], Y, scatter_kws={"color": "yellow"}, line_kws={"color": "red"})

plt.subplot(1, 3, 3)
plt.title("Incedent Rate")
sns.regplot(X['incident_rate'], Y, scatter_kws={"color": "blue"}, line_kws={"color": "red"})

```

**Multicollinearity**

```python
def calculate_vif(df):
    vif = pd.DataFrame()
    vif['features']=df.columns
    vif["vif_value"]=[variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif
```

```python
features = final_df.iloc[:, :-1]
calculate_vif(features)
```

```python
from sklearn.model_selection import train_test_split

x = final_df[['avg_count','avg_peryear','incident_rate']]
y = final_df[['target_death']]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=40, train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict((x_test))
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import *
import sklearn.metrics as metrics

mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
print(mae, mse)
```

```python
r2 = metrics.r2_score(y_test, y_pred)
print(r2)
```
